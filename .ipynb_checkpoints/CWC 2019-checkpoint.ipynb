```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
result=pd.read_csv('results.csv')
icc_ranking=pd.read_csv('icc_rankings.csv')
fixtures=pd.read_csv('fixtures.csv')
wc_data=pd.read_csv('World Cup 2019 Dataset.csv')
```

```python
wc_data.rename(columns={'Previous \r\nappearances':'previous_appearances', 'Previous \r\ntitles':'previous_titles',
       'Previous\r\n finals':'previous_finals', 'Previous\r\n semifinals':'previous_semifinals', 'Current \r rank':'Position'}, inplace=True)
```

```python
for i in result.columns:
    print(result[i].isnull().value_counts(), end='\n\n\n')
```

```python
wc_data_1=wc_data.copy()
wc_data_2=wc_data.copy()
```

```python
wc_data_1=pd.merge(wc_data_1,icc_ranking)
wc_data_2=pd.merge(wc_data_2,icc_ranking)
```

```python
wc_data_1.rename(columns={'Team':'Team_1' }, inplace=True)
wc_data_2.rename(columns={'Team':'Team_2' }, inplace=True)
```

```python
wc_data_1.head()
```

```python
df=pd.merge(pd.merge(result, wc_data_1, on='Team_1'), wc_data_2, on='Team_2',  suffixes=('_team_1', '_team_2'))
```

```python
df.head(10)
```

```python
df.to_csv('data.csv')
```
